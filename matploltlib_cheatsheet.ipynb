**MATPLOTLIB**

```python
import matplotlib.pyplot as plt
import numpy as np
```

***PIE CHART***

```python
x = ["A","B","C","D"]
y=np.linspace(50,100,4)

plt.pie(y , labels=x , autopct="%0.2f%%") #autopct for displaying percentage
plt.show()

```

***SCATTER PLOT***

```python
x = np.arange(1,11)
y= np.random.random(10)

plt.scatter(x,y)

plt.xlabel("num")
plt.ylabel("value")
plt.title("Test")
```

***2D PLOT***

```python
x = np.arange(1,21)
y= np.random.random(20)
z=np.random.random(20)
p=np.random.random(20)

#you can plot plot multiple graph in single one
plt.plot(x,y,color='red',marker="*",markersize=5,label="y")

plt.plot(x,z,color='blue',marker="+",markersize=5,label="z")

plt.plot(x,p,color='yellow',marker="D",markersize=5,label="p")

plt.legend() #this should be called to specipy labes of plots
plt.xlabel("num")
plt.ylabel("value")
plt.title("Test")
```

***BAR GRAPH***

```python
year = np.arange(2000,2011)
bike_price = np.random.randint(50000,100000,11)

plt.bar(year,bike_price,color="blue")

plt.xlabel("Year")
plt.ylabel("Bike Price")
plt.title("Bike Price Info")
plt.show()
```
